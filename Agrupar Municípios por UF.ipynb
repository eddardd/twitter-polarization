```python
import pandas as pd
```

```python
df = pd.read_csv('./data/Eleicoes2018SegundoTurno.csv')
```

```python
df
```

```python
sub_df = df.groupby(by='uf').sum()[['votosHaddad', 'votosBolsonaro']]
sub_df['totalVotos'] = sub_df['votosHaddad'] + sub_df['votosBolsonaro']
sub_df['percentHaddad'] = sub_df['votosHaddad'] / sub_df['totalVotos']
sub_df['percentBolsonaro'] = sub_df['votosBolsonaro'] / sub_df['totalVotos']
sub_df['numeroCandidatoMaisVotado'] = [
    13 if sub_df.values[i, -2] > sub_df.values[i, -1] else 17 for i in range(len(sub_df.values))
]

sub_df
```

```python
sub_df.to_csv('./data/ResultadoUFEleicoes2018.csv')
```
